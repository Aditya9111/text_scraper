```python
data=[]
with open('iisc_urls.csv',mode='r',encoding='utf-8-sig') as file:
  csv_reader = csv.DictReader(file)
  for csv_row in csv_reader:
    data.append({"domain":csv_row['domain'],"topic":csv_row['topic'],"url":csv_row['url']})

#print(urls)
## initializing the UserAgent object
user_agent = UserAgent()





## starting the loop
for url in data:
    ## getting the reponse from the page using get method of requests module
    page = requests.get(url['url'], headers={"user-agent": user_agent.chrome})

    ## storing the content of the page in a variable
    html = page.content
    page = requests.get(url['url'])
    textData=[]
    ## creating BeautifulSoup object
    soup = bs4.BeautifulSoup(page.content, "html.parser")
    for i in  range(len(soup.find_all('p'))):
      texts = soup.find_all('p')[i].get_text()
      if len(texts)>0:
        textData.append(texts);

    

    #print(textData)
    # model data with link,topic,domain
    final_data={'link':url['url'],'text':textData,'topic':url['topic'] ,'domain':url['domain']} 
    print(final_data)
    
    d_path = '/content/gdrive/My Drive/IISC_TEST/{}'.format(url['domain'])
    # create folder if not exists
    os.makedirs(d_path,exist_ok=True)

    # address to dump file
    gdrive_path = '/content/gdrive/My Drive/IISC_TEST/{}/{}-'.format(url['domain'],url['topic'])
    #print(gdrive_path)

    with open(f'{gdrive_path}'+str(time.time())+'.json', 'w') as f: 
       json.dump(final_data, f)

    
```
